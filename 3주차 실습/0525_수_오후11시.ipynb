## concat 메서드  
concat 메서드는 파라미터로 전달되는 값들을 기존 배열에 합쳐서 새로운 배열을 반환해주는 배열의 메서드들 중 하나이다. 한 가지 특별한 점은, 만약 파라미터로 전달되는 인자 값이 배열일 경우에는 해당 배열이 펼쳐지면서 기존 배열에 합쳐진다. 참고로, concat은 concatenation이라는 단어의 줄임말이라고 한다.  
pd.concat(  
* objs,  <span style='color:blue'>#Series, DataFrame, Panel object</span>  
* axis=0, <span style='color:blue'>#0: 위+아래로 합치기, 1: 왼쪽+오른쪽으로 합치기</span>  
* join='outer', <span style='color:blue'>#'outer': 합집합(union), 'inner': 교집합(intersection)</span>  
* ignore_index=False,  <span style='color:blue'>#False: 기존 index 유지, True: 기존 index 무시</span>  
* keys=None, <span style='color:blue'>#계층적 index 사용하려면 keys 튜플 입력</span>  
* levels=None,  
* names=None, <span style='color:blue'>#index의 이름 부여하려면 names 튜플 입력</span>  
* verify_integrity=False, <span style='color:blue'>#True: index 중복 확인</span>  
* copy=True) <span style='color:blue'>#복사</span>

## 파이썬에서는 0행, 1은 열 axis는 축

```python
#열 방향으로 데이터를 연결하는 경우
col_concat_1 = pd.concat([df1,df2,df3],axis = 1)
```

```python
col_concat_1
```

```python
#같은 열 이름이 있는 경우
col_concat_1['A']
```

```python
col_concat_1['NEW'] = ['n1','n2','n3','n4']
```

```python
col_concat_1
```

```python
#인덱스를 ignore_index(기존의 인덱스를 무시하고 싶을 때)
pd.concat([df1,df2,df3],axis=1,ignore_index=True)
```

### 공통의 열만 골라 연결하면 누락값이 생기지 않는다.

```python
#공통열을 가지고 오는 방법
pd.concat([df1,df2,df3],join='inner')
```

```python
print(df1)
print('---------------')
print(df3)
```

```python
pd.concat([df1,df3],
          ignore_index=False,
          join='inner')
```

```python
pd.concat([df1,df3],
         ignore_index=True,
         join='inner')
```

```python
df1.index = [0,1,2,3]
df2.index = [4,5,6,7]
df3.index = [0,2,5,7]
```

```python
print(df1)
print('----------------------')
print(df2)
print('----------------------')
print(df3)
```

```python
pd.concat([df1,df2,df3],axis=1)
```

```python
#공통의 행만 골라서 연결하면 된다.
pd.concat([df1,df3],axis=1,join='inner')
```

# Merge

- 기본적으로 내부 조인을 실행하며 메서드를 사용한 데이터프레임을 왼쪽으로 지정하고(기준으로 하고) 첫 번째 인자값으로 지정한 
데이터프레임을 오른쪽으로 지정한다.
- left_on, right_on이라는 인자값이 일치해야 한다.

```python
site = pd.read_csv('survey_site.csv')
survey = pd.read_csv('survey.csv')
visited = pd.read_csv('survey_visited.csv')
```

```python
site
```

```python
survey
```

```python
site
```

```python
visited
```

```python
#merge
m2o_merge = site.merge(visited, left_on='name',right_on='site')
```

```python
m2o_merge
```
