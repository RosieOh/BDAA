```python
#미국 표준 시간대 데이터 사용
import json
path = 'bitly_usagov/example.txt'
records = [json.loads(line) for line in open(path)]
```

```python
records[0]
```

```python
time_zones = [rec['tz'] for rec in records if 'tz' in rec]
time_zones[:10]
```

```python
def get_counts(sequence):
    counts = {}
    for x in sequence:
        if x in counts:
            counts[x] += 1
        else:
            counts[x] = 1
    return counts
```

```python
from collections import defaultdict

def get_counts2(sequence):
    counts = defaultdict(int) # 값이 0으로 초기화 됨
    for x in sequence:
        counts[x] += 1
    return counts
```

```python
counts = get_counts(time_zones)
counts['America/New_York']
```

```python
len(time_zones)
```

```python
def top_counts(count_dict, n = 10):
    value_key_pairs = [(count, tz) for tz, count in count_dict.items()]
    value_key_pairs.sort()
    return value_key_pairs[-n:]
```

```python
top_counts(counts)
```

```python
from collections import Counter
counts = Counter(time_zones)
counts.most_common(10)
```

```python
#히트맵 그리기 예제 
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = [10, 8]
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
df = flights.pivot('month', 'year', 'passengers')
df.head()
```

```python
# matplotlib 버전 히트맵
plt.pcolor(df)
plt.xticks(np.arange(0.5, len(df.columns), 1), df.columns)
plt.yticks(np.arange(0.5, len(df.index), 1), df.index)
plt.title('Heatmap by plt.pcolor()', fontsize=20)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Month', fontsize=14)
plt.colorbar()
plt.show()
```

```python
# seaborn 버전 히트맵
ax = sns.heatmap(df)
plt.title('Heatmap of Flight by seaborn', fontsize=20)
plt.show()
```

```python
# annotate each cell with the numeric value of integer format
sns.heatmap(df, annot=True, fmt='d')
plt.title('Annoteat cell with numeric value', fontsize=20)
plt.show()
```

```python
# 히트맵 색깔바꾸기 1.
sns.heatmap(df, cmap='RdYlGn_r')
plt.title('colormap of cmap=RdYlGn_r', fontsize=20)
plt.show()
```

```python
# 히트맵 색깔바꾸기 2.
sns.heatmap(df, cmap = 'YlGnBu')
plt.title('colormap of camp = YlGnBu', fontsize=20)
plt.show()
```

```python
# 히트맵 그리기 3.
df.style.background_gradient(cmap='summer')
```
